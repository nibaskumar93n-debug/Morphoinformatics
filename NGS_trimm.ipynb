```python
## Trimomatic install
!apt-get update -qq
!apt-get install -y openjdk-11-jre-headless
!wget -q https://github.com/usadellab/Trimmomatic/releases/download/v0.39/trimmomatic-0.39.zip
!unzip -q trimmomatic-0.39.zip
!java -jar Trimmomatic-0.39/trimmomatic-0.39.jar -version
```

```python
# Step 0: Define directories and variables  (best way)
# -------------------------------
INPUT_DIR="/content/raw_fastq"
OUTPUT_DIR="/content/trimmed_fastq"
ADAPTERS="Trimmomatic-0.39/adapters/TruSeq3-PE.fa"
THREADS=4

# Raw FASTQ filenames
R1="SRR31264075_1.fastq.gz"
R2="SRR31264075_2.fastq.gz"

# Output filenames automatically derived from input names
PAIRED1=f"{R1.replace('_1.fastq.gz','_1_paired.fq.gz')}"
UNPAIRED1=f"{R1.replace('_1.fastq.gz','_1_unpaired.fq.gz')}"
PAIRED2=f"{R2.replace('_2.fastq.gz','_2_paired.fq.gz')}"
UNPAIRED2=f"{R2.replace('_2.fastq.gz','_2_unpaired.fq.gz')}"

# Create output directory if it doesn't exist
import os
os.makedirs(OUTPUT_DIR, exist_ok=True)
# -------------------------------
# Step 1: Run Trimmomatic PE
# -------------------------------
!java -jar Trimmomatic-0.39/trimmomatic-0.39.jar PE \
  -threads {THREADS} \
  {INPUT_DIR}/{R1} {INPUT_DIR}/{R2} \
  {OUTPUT_DIR}/{PAIRED1} {OUTPUT_DIR}/{UNPAIRED1} \
  {OUTPUT_DIR}/{PAIRED2} {OUTPUT_DIR}/{UNPAIRED2} \
  ILLUMINACLIP:{ADAPTERS}:2:30:10 \
  LEADING:3 TRAILING:3 SLIDINGWINDOW:4:15 MINLEN:36
```
