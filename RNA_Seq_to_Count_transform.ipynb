```python
!hisat2-build Homo_sapiens.GRCh38.dna.primary_assembly.fa human_index
```

```python
!mkdir -p trimmed_reads
```

```python
%%bash

mkdir -p trimmed_reads

for sample in SRR35777807 SRR35777810
do
    trimmomatic PE \
    -threads 4 \
    /content/fastq_files/${sample}_1.fastq.gz \
    /content/fastq_files/${sample}_2.fastq.gz \
    trimmed_reads/${sample}_1_paired.fastq.gz \
    trimmed_reads/${sample}_1_unpaired.fastq.gz \
    trimmed_reads/${sample}_2_paired.fastq.gz \
    trimmed_reads/${sample}_2_unpaired.fastq.gz \
    ILLUMINACLIP:TruSeq3-PE.fa:2:30:10 \
    LEADING:3 \
    TRAILING:3 \
    SLIDINGWINDOW:4:15 \
    MINLEN:36
done
```

```python
%%bash

for sample in SRR35777807 SRR35777810
do
    hisat2 -p 4 \
    -x human_index \
    -1 trimmed_reads/${sample}_1_paired.fastq.gz \
    -2 trimmed_reads/${sample}_2_paired.fastq.gz \
    -S ${sample}.sam
done
```

```python
%%bash

for sample in SRR35777807 SRR35777810
do
    samtools view -bS ${sample}.sam | \
    samtools sort -o ${sample}.sorted.bam
done
```

```python
%%bash

for sample in SRR35777807 SRR35777810
do
    samtools index ${sample}.sorted.bam
done
```

```python
!featureCounts \
-T 4 \
-p \
-a Homo_sapiens.GRCh38.111.gtf \
-o gene_counts.txt \
SRR35777807.sorted.bam \
SRR35777810.sorted.bam
```
