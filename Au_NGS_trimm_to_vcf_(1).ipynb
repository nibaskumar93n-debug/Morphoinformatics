```python
import os

# 1. Configuration - Change only these
sra_id = "SRR10568744"
threads = 4

# 2. Path Setup
ref_genome = "/content/drive/MyDrive/reference/hg38.fa"
input_dir = f"/content/drive/MyDrive/{sra_id}" # Based on your previous path
output_bam = f"/content/drive/MyDrive/{sra_id}.sorted.bam"

# 3. File Verification
r1 = f"{input_dir}/{sra_id}_1_paired.fq.gz"
r2 = f"{input_dir}/{sra_id}_2_paired.fq.gz"

# Check if the reference index exists (bwa needs .amb, .ann, .bwt, .pac, .sa)
if not os.path.exists(f"{ref_genome}.bwt"):
    print(f"Warning: Reference index not found. Indexing now...")
    !bwa index {ref_genome}

# 4. Execution
print(f"--- Starting Alignment for {sra_id} ---")

# We use the \ to break lines for readability
!bwa mem -t {threads} \
  -R "@RG\tID:{sra_id}\tSM:{sra_id}\tPL:ILLUMINA" \
  {ref_genome} \
  {r1} \
  {r2} \
| samtools view -@ {threads} -Sb - \
| samtools sort -@ {threads} -o {output_bam}

# 5. Final Step: Index the BAM file
print(f"--- Alignment Finished. Creating BAM Index (.bai) ---")
!samtools index {output_bam}

print(f"Done! Your sorted BAM and Index are at: {output_bam}")
!ls -lh {output_bam}*
```

```python
import os

# 1. Configuration
sra_id = "SRR10568744"
gatk_path = "/content/gatk-4.3.0.0/gatk"
java_opts = "-Xmx8g"

# 2. Path Setup
ref = "/content/drive/MyDrive/reference/hg38.fa"
input_bam = f"/content/drive/MyDrive/{sra_id}.dedup.bam"
recal_table = f"/content/drive/MyDrive/{sra_id}.recal.table"
output_bam = f"/content/drive/MyDrive/{sra_id}.recal.bam"

# Known Sites (Ensure these paths are correct in your Drive)
known_snps = "/content/drive/MyDrive/known_sites/1000G_phase1.snps.high_confidence.hg38.vcf.gz"
known_indels = "/content/drive/MyDrive/known_sites/Mills_and_1000G_gold_standard.indels.hg38.vcf.gz"

# 3. Create output directory for tables
os.makedirs("/content/drive/MyDrive/", exist_ok=True)

# 4. Step A: BaseRecalibrator (Generates the .table)
print(f"--- Step A: Generating Recalibration Table for {sra_id} ---")
!{gatk_path} --java-options "{java_opts}" BaseRecalibrator \
  -I {input_bam} \
  -R {ref} \
  --known-sites {known_snps} \
  --known-sites {known_indels} \
  -O {recal_table}

# 5. Step B: ApplyBQSR (Creates the final recalibrated BAM)
print(f"\n--- Step B: Applying BQSR to create final BAM for {sra_id} ---")
if os.path.exists(recal_table):
    !{gatk_path} --java-options "{java_opts}" ApplyBQSR \
      -R {ref} \
      -I {input_bam} \
      --bqsr-recal-file {recal_table} \
      -O {output_bam}
    print(f"✅ BQSR Complete! Final file: {output_bam}")
else:
    print("❌ ERROR: Recalibration table was not created.")
```

```python
import os

# 1. Configuration - Set your Tumor and Normal IDs
tumor_id = "SRR10568744"
normal_id = "SRR10568745"
gatk_path = "/content/gatk-4.3.0.0/gatk"
java_opts = "-Xmx8g"

# 2. Path Setup
ref = "/content/drive/MyDrive/reference/hg38.fa"
recal_dir = "/content/drive/MyDrive"

# Input BAMs
tumor_bam = f"{recal_dir}/{tumor_id}.recal.bam"
normal_bam = f"{recal_dir}/{normal_id}.recal.bam"

# Resources
germline_resource = "/content/drive/MyDrive/reference/af-only-gnomad.hg38.vcf.gz"
pon = "/content/drive/MyDrive/reference/1000g_pon.hg38.vcf.gz"

# Output
output_vcf = f"{recal_dir}/{tumor_id}_vs_{normal_id}.unfiltered.vcf.gz"

# 3. Execution: Mutect2
print(f"--- Running Mutect2 Somatic Calling ---")
print(f"Tumor: {tumor_id} | Normal: {normal_id}")

if os.path.exists(tumor_bam) and os.path.exists(normal_bam):
    !{gatk_path} --java-options "{java_opts}" Mutect2 \
      -R {ref} \
      -I {tumor_bam} \
      -I {normal_bam} \
      -tumor {tumor_id} \
      -normal {normal_id} \
      --germline-resource {germline_resource} \
      --panel-of-normals {pon} \
      -O {output_vcf}

    print(f"\n✅ Success! Unfiltered VCF created: {output_vcf}")
else:
    print(f"❌ ERROR: One or both BAM files are missing in {recal_dir}")
```

```python
import os
from google.colab import drive

# 0. Mount Google Drive (Required if not already mounted)
if not os.path.exists('/content/drive'):
    drive.mount('/content/drive')

# 1. Configuration
tumor_id = "SRR10568744"
normal_id = "SRR10568745"
# This matches your file: SRR36736757_vs_SRR36736758.unfiltered.vcf.gz
pair_id = f"{tumor_id}_vs_{normal_id}"

gatk_path = "/content/gatk-4.3.0.0/gatk"
java_opts = "-Xmx8g"

# 2. Path Setup
ref = "/content/drive/MyDrive/reference/hg38.fa"
recal_dir = "/content/drive/MyDrive"

# Input/Output Files
# Using "/" to separate directory from the filename
unfiltered_vcf = f"{recal_dir}/{pair_id}.unfiltered.vcf.gz"
stats_file = f"{unfiltered_vcf}.stats"
filtered_vcf = f"{recal_dir}/{pair_id}.filtered.vcf.gz"

# 3. Execution: FilterMutectCalls
print(f"--- Filtering Somatic Variants for {pair_id} ---")
print(f"Targeting VCF: {unfiltered_vcf}")
print(f"Targeting Stats: {stats_file}")

# Verify files exist before running GATK
if os.path.exists(unfiltered_vcf) and os.path.exists(stats_file):
    !{gatk_path} --java-options "{java_opts}" FilterMutectCalls \
      -R {ref} \
      -V {unfiltered_vcf} \
      --stats {stats_file} \
      -O {filtered_vcf}

    print(f"\n✅ Filtering Complete!")
    print(f"Final Somatic VCF: {filtered_vcf}")
else:
    print(f"\n❌ ERROR: File not found.")
    if not os.path.exists(unfiltered_vcf):
        print(f"Missing VCF: {unfiltered_vcf}")
    if not os.path.exists(stats_file):
        print(f"Missing Stats: {stats_file}")
```
