```python
# Upload the Gym data in google drive
gym = pd.read_csv('/content/drive/MyDrive/datasets/gym_membership.csv')
# Display last 5 rows
print(gym.tail())
# Check data types
print(gym.dtypes)
```

```python
# Task 1: Basic Matplotlib (3 plots)
# 1. Bar chart of gym membership types
# Use: gym['abonoment_type'].value_counts()
gym_bar=gym['abonoment_type'].value_counts()
plt.bar(gym_bar.index, gym_bar.values, color=['red', 'blue'])
plt.title('Abonoment Type')
plt.xlabel('Types')
plt.ylabel('counts')
plt.grid(True, which="both", linestyle="-", linewidth=0.7, alpha=0.7)
plt.savefig("/content/drive/MyDrive/datasets/bar_gym_type.png",dpi=300, bbox_inches='tight')
plt.show()
# 2. Histogram of time in gym
# Use: plt.hist(gym['avg_time_in_gym'])
plt.hist(gym['avg_time_in_gym'],bins=10,color= 'green')
plt.title('Average Time')
plt.xlabel('time')
plt.ylabel('counts')
plt.grid(True, which="both", linestyle="-", linewidth=0.7, alpha=0.7)
plt.savefig("/content/drive/MyDrive/datasets/avg_time_in_gym.png",dpi=300, bbox_inches='tight')
plt.show()
# 3. Scatter plot: Age vs visits per week
# Use: plt.scatter(gym['Age'], gym['visit_per_week'])
plt.scatter( gym['Age'], gym['visit_per_week'])
plt.grid(True,which="both", linestyle="-", linewidth=0.7, alpha=0.7)
plt.xlabel('Age')
plt.ylabel('Visit per week')
plt.title('Age vs visit per week')
plt.savefig("/content/drive/MyDrive/datasets/Age_visit_per_week.png",dpi=300, bbox_inches='tight')
plt.show()
```

```python
# 1. Box plot: Gender vs Age
# Use: sns.boxplot(data=gym, x='gender', y='Age')
sns.boxplot(data=gym, x='gender', y='Age')
plt.title('Gender vs Age')
plt.grid(True, which="both", linestyle="-", linewidth=0.7, alpha=0.7)
plt.savefig("/content/drive/MyDrive/datasets/seaborn_gender_age.png",dpi=300, bbox_inches='tight')
plt.show()
# 2. Count plot: Days per week
# Use: sns.countplot(data=gym, x='days_per_week')
plt.figure(figsize=(10,8))
sns.countplot(data=gym, x='days_per_week')
plt.title('Days_per_week')
plt.xlabel('days_per_week')
plt.ylabel('counts')
plt.xticks(size=6,rotation=45,ha='right',fontsize=9)
plt.grid(True, which="both", linestyle="-", linewidth=0.7, alpha=0.7)
plt.savefig("/content/drive/MyDrive/datasets/seaborn_days_per_week.png",dpi=300, bbox_inches='tight')
plt.show()
# 3. Bar plot: Average visits by membership type
# Use: sns.barplot(data=gym, x='abonoment_type', y='visit_per_week')
sns.barplot(data=gym, x='abonoment_type', y='visit_per_week')
plt.title('Average visit by membership type')
plt.grid(True, which="both", linestyle="-", linewidth=0.7, alpha=0.7)
plt.savefig("/content/drive/MyDrive/datasets/seaborn_visit_membership_type.png",dpi=300, bbox_inches='tight')
plt.show()
```
