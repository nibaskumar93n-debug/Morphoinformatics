```python
import os
import pandas as pd

# Set your data directory (adjust if different)
data_dir = "/content/raw_fastq2"  # or wherever you unzipped

# Get all _1.fastq.gz files (forward reads)
all_files = os.listdir(data_dir)
r1_files = sorted([f for f in all_files if '_1.fastq.gz' in f])

print(f"Found {len(r1_files)} samples")

# Create manifest
manifest_data = []
for r1 in r1_files:
    # Extract sample ID (e.g., SRR10604964 from SRR10604964_1.fastq.gz)
    sample_id = r1.replace('_1.fastq.gz', '')
    # Create corresponding R2 filename
    r2 = r1.replace('_1.fastq.gz', '_2.fastq.gz')
    # Add to manifest
    manifest_data.append({
        'sample-id': sample_id,
        'forward-absolute-filepath': os.path.join(data_dir, r1),
        'reverse-absolute-filepath': os.path.join(data_dir, r2)
    })

# Create dataframe and save
df = pd.DataFrame(manifest_data)
df.to_csv('manifest.tsv', sep='\t', index=False)

print(f"✓ Manifest created with {len(manifest_data)} samples!")
print("\nFirst few rows:")
print(df.head())
print("\nLast few rows:")
print(df.tail())
```

```python
# Activate QIIME2 environment and run import
!source /usr/local/etc/profile.d/conda.sh && conda activate qiime2-amplicon-2024.10 && qiime tools import \
  --type 'SampleData[PairedEndSequencesWithQuality]' \
  --input-path manifest.tsv \
  --input-format PairedEndFastqManifestPhred33V2 \
  --output-path demux-paired-end.qza
```

```python
!source /usr/local/etc/profile.d/conda.sh && conda activate qiime2-amplicon-2024.10 && qiime demux summarize \
  --i-data demux-paired-end.qza \
  --o-visualization demux-summary.qzv
```

```python
!source /usr/local/etc/profile.d/conda.sh && conda activate qiime2-amplicon-2024.10 && qiime dada2 denoise-paired \
  --i-demultiplexed-seqs demux-paired-end.qza \
  --p-trim-left-f 3 \
  --p-trim-left-r 3 \
  --p-trunc-len-f 145 \
  --p-trunc-len-r 135 \
  --o-table table.qza \
  --o-representative-sequences rep-seqs.qza \
  --o-denoising-stats denoising-stats.qza
```
